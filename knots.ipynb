The command below makes a random knot. Each coordinate is a Fourier polynomial of degree given by the second argument (in this case, 60), with coefficients gaussian, with mean 0 and standard deviation of the nth coefficient decaying as $n^{-1.4}$ (or whatever the first argument was).

````python
ff = make_rand_knot(1.4, 60)
````

This makes a tube around the knot, and visualizes it using ```pyvista```

````python
thetube = make_tube(ff, colab = False)
````

Now we can compute the Alexander polynomial of our random knot.

````python
pp = get_alex_poly(ff)
````
